```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../../Datasets/census.csv")
df.head()
```

```python
import timeit
```

```python
%%timeit -n 3

for state in df["STATE"].unique():
    avg = np.average(df.where(df["STATE"] == state).dropna()["CENSUS2010POP"])
    print("The counties in state "+str(state)+ " have the average population of "+str(avg))
```

```python
%%timeit -n 3

for group,frame in df.groupby("STATE"):
    avg = np.average(frame["CENSUS2010POP"])
    print("the counties in state "+str(group)+" having the population of "+str(avg))
    
```

```python
df = pd.read_csv("../../Datasets/census.csv")
df.set_index("STATE")

def set_batch_number(item):
    if item < 30:
        return 0
    if item < 57:
        return 1
    return 2

for group,frame in df.groupby(set_batch_number):
    print("The group "+str(group)+" has the count of "+str(len(frame)))
```
