```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../../Datasets/census.csv")
df.head()
```

```python
import timeit
```

```python
%%timeit -n 3

for state in df["STATE"].unique():
    avg = np.average(df.where(df["STATE"] == state).dropna()["CENSUS2010POP"])
    print("The counties in state "+str(state)+ " have the average population of "+str(avg))
```

```python
%%timeit -n 3

for group,frame in df.groupby("STATE"):
    avg = np.average(frame["CENSUS2010POP"])
    print("the counties in state "+str(group)+" having the population of "+str(avg))
    
```

```python
df = pd.read_csv("../../Datasets/census.csv")
df.set_index("STATE")

def set_batch_number(item):
    if item < 30:
        return 0
    if item < 57:
        return 1
    return 2

for group,frame in df.groupby(set_batch_number):
    print("The group "+str(group)+" has the count of "+str(len(frame)))
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("../../Datasets/listings.csv")
df.head()
```

```python
df=df.set_index(["cancellation_policy","review_scores_value"])

for group,frame in df.groupby(level=(0,1)):
    print(group)
```

```python
def grouping_fun(item):
    if item[1] == 10.0:
        return (item[0],'10.0')
    else:
        return (item[0],"not 10.0") 
    
for group,frame in df.groupby(grouping_fun):
    print(group)
```
